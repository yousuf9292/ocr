```python
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
reader=easyocr.Reader(lang_list=['ar'])
```

```python
img=cv2.imread("C:\\Users\\yousuf\\Desktop\\fatwa.jpg")
```

```python
plt.imshow(img)
```

```python
y_top=450
y_bottom=500
x_left=800
x_right=950
```

```python
plt.imshow(img[y_top:y_bottom,x_left:x_right])
```

```python
single_word=img[y_top:y_bottom,x_left:x_right]
```

```python
output=reader.readtext(single_word,detail=0)
```
